```python
# Changing datatype to the string
df['Qty'] = df['Qty'].astype(str)
```

```python
# creating dictionary to change the value in the correct form
word_to_number = {
    'One': '1',
    'Two': '2'
}
# Replacing the values in the dataset
df['Qty'] = df['Qty'].replace(word_to_number)
```

```python
# checking the output those value change or not
df.Qty.value_counts()
```

```python
shorttofull = {
    'W' : 'Women',
    'M' : 'Men'
}

df['Gender'] = df['Gender'].replace(shorttofull)
```

```python
# converting the age into the bins

labels = ['{0} - {1}'.format(i,i+19) for i in range(1,78,20)]
bins = list(range(1, 82, 20))
df['Age_Group'] = pd.cut(df.Age,bins = bins,right=False,labels=labels)
```

```python
df.Age_Group.value_counts()
```
